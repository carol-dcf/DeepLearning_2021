## Plot Losses CycleGAN

```python
import matplotlib.pyplot as plt
```

```python
def plot_losses(log_file):
    file = open(log_file, 'r')
    lines = file.readlines()

    D_A = []
    G_A = []
    cycle_A = []
    idt_A = []
    D_B = []
    G_B = []
    cycle_B = []
    idt_B = []

    cur_epoch = 1
    for i in range(len(lines)):
        line = lines[i]

        if line[0] != '(':
            continue

        part = line.split()

        epoch = int(part[1][:-1])
        if epoch > cur_epoch:

            cur_epoch = epoch
            last_line = lines[i-1]
            last_line_part = last_line.split()
            try:
                D_A.append(float(last_line_part[9]))
                G_A.append(float(last_line_part[11]))
                cycle_A.append(float(last_line_part[13]))
                idt_A.append(float(last_line_part[15]))
                D_B.append(float(last_line_part[17]))
                G_B.append(float(last_line_part[19]))
                cycle_B.append(float(last_line_part[21]))
                idt_B.append(float(last_line_part[23]))
            except:
                continue
    
    plt.figure(0)
    plt.plot(D_A)
    plt.plot(G_A)
    plt.legend(['D_A', 'G_A'])
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.title('Cartoons generator and its discriminator')
    plt.grid()
    
    plt.figure(1)
    plt.plot(D_B)
    plt.plot(G_B)
    plt.legend(['D_B', 'G_B'])
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.title('Face generator and its discriminator')
    plt.grid()
    
    plt.figure(2)
    plt.plot(cycle_A)
    plt.plot(cycle_B)
    plt.legend(['cycle_A', 'cycle_B'])
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.title('Forwards and Backwards consistency loss')
    plt.grid()
    
    plt.figure(3)
    plt.plot(idt_A)
    plt.plot(idt_B)
    plt.legend(['idt_A', 'idt_B'])
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.title('Identity losses')
    plt.grid()
```
